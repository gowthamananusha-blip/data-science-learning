```python
import numpy as np

# Roll a dice 10,000 times
rolls = np.random.randint(1, 7, size=10000)

# Calculate what we need
expected_value = np.mean(rolls)  # average
variance = np.var(rolls)          # spread

# Print results
print("=" * 50)
print("DICE SIMULATION RESULTS")
print("=" * 50)
print(f"Expected Value (Average): {expected_value}")
print(f"Variance: {variance}")
print("\nProbabilities for each number:")
print("-" * 30)

# Calculate probability for each number (1 to 6)
for number in range(1, 7):
    count = np.sum(rolls == number)
    probability = count / 10000
    print(f"Probability of {number}: {probability:.2%}")
```
